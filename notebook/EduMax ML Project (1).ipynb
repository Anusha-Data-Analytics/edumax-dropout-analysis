## $\color{brown}{Retention Analysis}$ 

```python
# Dealing with Users with 0 Session
enrolled_user = enrollment_data["user_id"].unique()

session_user = session_data["user_id"].unique()

user_zero_session = []

for user in enrolled_user:
    if user not in session_user:
        user_zero_session.append(user)

print(f"Out of {len(enrolled_user)} users, {len(user_zero_session)} have Zero Session  ")
```

##### Most users at least try the platform — dropout happens after engagement starts

```python
# Finding How many people are dropped among Zero Session Users
dropped_user= list(enrollment_data["user_id"][enrollment_data["status"]=="dropped"])
len_zero_session = len(user_zero_session)
len_zero_session_dropped = len(set(dropped_user) & set(user_zero_session))
print(f"Out of {len_zero_session}, {len_zero_session_dropped} are dropped out!")

# finding Dropout rate of Zero Session User 
zero_session_dropped = round(len_zero_session_dropped/len_zero_session *100, 2)
print("Dropped Out Ratio of User who never start Session: ", zero_session_dropped,"%")
```

##### <font color="dark green">  Users with *Zero Sessions* show ~78% dropout rate, indicating a critical onboarding gap, though they represent a small portion of total users.

```python
# Creating User level Data
user_level_data = enrollment_data[["user_id","status"]]

#grouping by user_id
user_level_data = user_level_data.groupby("user_id")["status"].apply(list).reset_index()

#function to decide final status of each user
def decide_status(x):
    if 'dropped' in x:
        return 'dropped'
    elif all(s=='completed' for s in x):
        return 'completed'
    else:
        return 'active'
        
user_level_data['final_status'] = user_level_data["status"].apply(decide_status)

# copying user level data with all status list for reference
user_level_data_copy = user_level_data.copy()

#getting only user with final status 
user_level_data = user_level_data[['user_id','final_status']]
print("Final User Level Data for Analysis:\n",user_level_data.head(10))

print("Dropped Users: ",(user_level_data["final_status"]=="dropped").sum())
print("Completed Users: ",(user_level_data["final_status"]=="completed").sum())
print("Active Users: ",(user_level_data["final_status"]=="active").sum())
```

### <font color="black">  Engagement Metrics 

```python
# Calculating Engagement metrix
engagement_metrix_data = session_data.groupby("user_id").agg(session_count_per_user=("user_id","size"),total_time_spent=("duration_minutes","sum"),avg_duration=("duration_minutes","mean")).reset_index()


#getting final data by merging zero session users with users with sessions
user_level_data = user_level_data.merge(engagement_metrix_data,how="left")
print("\n\nUser level Data with Session Count:\n",user_level_data.head())

# Replacing 'NaN' values in zero session users to 0
user_level_data['session_count_per_user'] = user_level_data['session_count_per_user'].fillna(0)
user_level_data['total_time_spent'] = user_level_data['total_time_spent'].fillna(0)
user_level_data['avg_duration'] = user_level_data['avg_duration'].fillna(0)

# Checking Zero Session users in final user level data, user_id (U1153) has no session checked in previously
print("\n\nChecking Zero Session users in final user level data : \n",user_level_data[user_level_data["user_id"]=="U1153"])

#confirming we have 9 zero session users as we checked earlier
print("No. of Users with zero sessions: ",(user_level_data['session_count_per_user']==0).sum())
```

```python
user_level_data['avg_duration'] = round(user_level_data['avg_duration'],2)

print("The Final Dataset for Retention Analysis and Prediction: \n\n",user_level_data.head(10))
```

```python
#Need to analyse data distribution in user level data to create buckets
print("User Level Data Distribution:\n",engagement_metrix_data.describe())

# Creating Buckets so that we analyse the dropput ratio
session_count_bucket = [[0,2,'0-2'],[3,4,'3-4'],[5,6,'5-6'],[7,float('inf'),'7+']]
total_time_spent_bucket = [[0,100,'0-100'],[101,200,'101-200'],[201,300,'201-300'],[301,float('inf'),'300+']]
avg_duration_bucket = [[0,25,'0-25'],[25,45,'25-45'],[45,65,'45-65'],[65,float('inf'),'65+']]

# Fuction to decide which bucket each row falls
def decide_bucket(value,bucket):
    for start,end,refer in bucket:
        if start<=value and value<=end:
            return refer
    return 'unknown'
    

# Adding user level data into each bucket
print("\n\nUser level Data after bucket allocation: ")
user_level_data["session_count_bucket"] = user_level_data["session_count_per_user"].apply(lambda x : decide_bucket(x,session_count_bucket))
user_level_data["total_time_spent_bucket"] = user_level_data["total_time_spent"].apply(lambda x: decide_bucket(x, total_time_spent_bucket))
user_level_data["avg_duration_bucket"] = user_level_data["avg_duration"].apply(lambda x: decide_bucket(x,avg_duration_bucket))
user_level_data.head()
```

```python
# Need to analyse value to status to find dropout ration : dropped = 1 , not_dropped=0

user_level_data['is_dropped'] = user_level_data["final_status"].apply(lambda x: 1 if x=='dropped' else 0)
print("User level data after defining dropout variable: \n",user_level_data.head(10))
```

### <font color="black">  Dropout Ratio

```python
# For session_count
session_count_dropout = user_level_data.groupby('session_count_bucket').agg(total_users=("user_id","count"),dropped_user=("is_dropped","sum")).reset_index()

session_count_dropout['dropout_ratio (%)'] = round((session_count_dropout['dropped_user']/session_count_dropout['total_users']*100),2)
print("Engagement Metrics Session Count Dropout Ratio Data:")
session_count_dropout.head()
```

##### <font color="dark green"> Users dropping out are not just low-engagement users. The highest dropout occurs at 3–4 sessions, indicating a critical retention gap after initial onboarding.

```python
# For Total_time_spent

total_time_dropout = user_level_data.groupby("total_time_spent_bucket").agg(total_users = ("user_id","count"),dropped_users = ("is_dropped","sum")).reset_index()
total_time_dropout["dropout_ratio (%)"] = round((total_time_dropout["dropped_users"]/total_time_dropout["total_users"]*100),2)

print("Engagement Metrics total Time Spent Dropout Ratio Data:")
total_time_dropout.head()
```

##### <font color="dark green">Users with moderate engagement (3–6 sessions or 100–300 minutes) show the highest dropout, indicating a retention failure after initial adoption but before long-term commitment

```python
# For Average duration

avg_duration_dropout = user_level_data.groupby("avg_duration_bucket").agg(total_users=("user_id","count"),dropped_users=("is_dropped","sum")).reset_index()
avg_duration_dropout["dropout_ratio (%)"] = round((avg_duration_dropout['dropped_users']/avg_duration_dropout['total_users']*100),2)

print("Engagement Metrics Average Duration Dropout Ratio Data:")
avg_duration_dropout.head(20)
```

##### <font color="dark green">While higher total engagement reduces dropout, users with longer average session durations exhibit higher dropout rates, indicating that binge-learning behavior negatively impacts retention.

### <font color="dark blue"> Final Core Insight:
##### <font color="purple"> Retention is driven by consistent engagement (more sessions and higher total time), while binge-like behavior (high average session duration) increases the likelihood of dropout.” 

### <font color="black">  Behavioral Segmentation

```python
# Funtion to decide the feature segment
def decide_category(value,threshold):
    for start,end,category  in threshold:
        if start<=value and value <=end:
            return category
    else:
        return 'unknown'
```

```python
# Defining Thresholds
session_count_threshold = [[0,2,'low'],[3,5,'medium'],[6,float('inf'),'high']]
total_time_threshold = [[0,100,'low'],[101,220,'medium'],[221,float('inf'),'high']]
avg_duration_threshold = [[0,25,'low'],[25,45,'medium'],[45,float('inf'),'high']]

user_feature_category = user_level_data[['user_id','session_count_per_user','total_time_spent','avg_duration']]

# Creating Feature Categories
user_feature_category["session_category"] = user_feature_category['session_count_per_user'].apply(lambda x: decide_category(x,session_count_threshold))
user_feature_category['time_category'] = user_feature_category['total_time_spent'].apply(lambda x: decide_category(x,total_time_threshold))
user_feature_category['duration_category'] = user_feature_category['avg_duration'].apply(lambda x: decide_category(x,avg_duration_threshold))

#user_feature_category['status'] = user_level_data['final_status']
user_feature_category = user_feature_category.merge(user_level_data[['user_id','final_status','is_dropped']],on='user_id')
                                                    
print("Feature Category Data : \n")
user_feature_category.head()

```

```python
# decinding segment
def decide_segment(user_data):  
    if user_data['session_category']=='high':
        return'Consistent User'
    elif user_data['duration_category']=='high' and user_data['session_category'] != 'high':
        return 'Binge User'
    elif user_data['session_category']=='low' and user_data['time_category']=='low':
        return 'Low Engagement'
    else:
        return 'Moderate Users'

# defining user behavioral segments
user_feature_category["user_segment"] = user_feature_category.apply(decide_segment,axis=1)
print("User Data with Behavioral Segments: \n")
user_feature_category.head()
```

```python
# finding dropout ratio of each behavioral segment 

behavioral_segment_dropout = user_feature_category.groupby('user_segment').agg(total_users=('user_segment','count'),dropped_users=('is_dropped','sum')).reset_index()
behavioral_segment_dropout['dropout_ratio (%)']=round((behavioral_segment_dropout['dropped_users']/behavioral_segment_dropout['total_users']*100),2)

print("Behavioural Segment Dropout Ratio: ")
behavioral_segment_dropout.head()
```

### <font color="dark blue">📊 Behavioural Segment Dropout Analysis

##### <font color="purple">📌 Key Insights

Binge Users exhibit the highest dropout rate (70%), indicating that users with high-intensity but low-frequency engagement are most prone to churn. In contrast, Consistent and Low Engagement users show relatively lower dropout rates (50–53%), highlighting the importance of regular learning behavior over total consumption.

##### <font color="purple">📌 Observations
Binge Users (69.66%) have the highest dropout among all segments 🚨
Moderate Users (56.08%) show unstable behavior with moderately high dropout
Consistent Users (53.41%) perform better, indicating regular engagement improves retention
Low Engagement Users (50.59%) surprisingly have the lowest dropout among all segments

##### <font color="purple">📌 Business Interpretation

The results suggest that engagement pattern matters more than engagement volume.
Binge users tend to consume large amounts of content in fewer sessions, which can lead to cognitive overload and burnout, increasing the likelihood of dropout.

On the other hand, Low Engagement users progress slowly but steadily, avoiding overload, which may contribute to relatively better retention. Consistent users, who engage regularly, demonstrate more stable learning behavior and improved retention outcomes.

##### <font color="purple">📌 Recommendations
- Convert Binge Users → Consistent Users
      Break content into smaller, manageable chunks
      Introduce daily learning goals and streak-based systems
      Implement reminders to encourage frequent engagement
- Stabilize Moderate Users
      Use nudges and personalized recommendations to build consistency
- Strengthen Consistent Users
      Introduce gamification, rewards, and progress tracking to maintain engagement
- Support Low Engagement Users
      Provide beginner-friendly pathways and gentle nudges to increase activity without overwhelming them

```python
# getting required columns from course data
course_seg_data = course_data[['category','course_id']]

# merging course segment data to enrollment data
course_seg_data = course_seg_data.merge(enrollment_data[['course_id','user_id','status']])
course_seg_data = course_seg_data.merge(user_age_segment_data[['user_id','age_bucket','user_segment']])

# making status numerical for calculation
course_seg_data['is_dropped'] = course_seg_data['status'].apply(lambda x: 1 if x=='dropped' else 0)

# Need to get course categories with priority one
status_priority = { 
    'dropped':2,
    'completed':1,
    'active':0
}
course_seg_data['status_rank'] = course_seg_data['status'].map(status_priority)
print("The Course Level Dropout Data: (before) \n\n",course_seg_data.head(),course_seg_data.shape)
course_seg_data = course_seg_data.sort_values(['user_id','course_id','status_rank'],ascending=[False,False,False]).drop_duplicates(subset=['user_id','course_id'])
course_seg_data = course_seg_data[course_seg_data['status']!='active']
course_seg_data = course_seg_data.drop(columns="status_rank")
print("The Course Level Dropout Data: (shape)\n\n",course_seg_data.shape)
print('\n\nDuplicates:\n',course_seg_data[course_seg_data.duplicated(['user_id','course_id'], keep=False)])
course_seg_data.head()
```

“All course enrollments where the user has reached a final outcome (either completed or dropped)”
